```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset("flights")
dataset
```

```python
sns.barplot(x="year",y="passengers",data = dataset)
plt.grid()
plt.show()
```

```python
sns.barplot(x="year",y="passengers",data = dataset,hue="month")
plt.grid()
plt.show()
```
