```python
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
Dataset = sns.load_dataset("planets")
Dataset
```

```python
Dataset = sns.load_dataset("planets").head(20)
sns.pointplot(x="mass",y="distance",data=Dataset)
plt.grid()
plt.show()
```

```python
Dataset = sns.load_dataset('planets').head(100)
sns.pointplot(x="mass",y="distance",data=Dataset,hue="method",linestyles="--")
plt.grid()
plt.show()
```
