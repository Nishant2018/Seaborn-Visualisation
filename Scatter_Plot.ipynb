```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dataset = sns.load_dataset("car_crashes")
dataset
```

```python
sns.scatterplot(x="speeding",y="total",data=dataset,hue="alcohol",marker=">")
plt.show()
```

```python
sns.scatterplot(x="speeding",y="total",data=dataset,hue="alcohol",marker="o",palette="husl")
plt.title("Supervised")
plt.show()
```

```python
import seaborn as sns
dataset = sns.load_dataset("car_crashes").head(25)
dataset
sns.set_style("whitegrid")
sns.scatterplot(x="speeding",y="total",data=dataset,hue="alcohol")
plt.show()
```

```python
sns.stripplot(x="speeding",y="total",data=dataset,marker="o",palette="deep",hue="alcohol")
plt.show()
```

```python
sns.swarmplot(x="speeding",y="total",data=dataset,marker=">",hue="alcohol")
plt.show()
```
