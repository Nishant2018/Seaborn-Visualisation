```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
Dataset = sns.load_dataset("penguins")
Dataset
```

```python
sns.countplot(x="island",data=Dataset)
plt.grid()
plt.show()
```

```python
sns.countplot(x="island",data=Dataset,palette="husl",hue="sex")
plt.grid()
plt.show()
```

```python
sns.countplot(x="species",data=Dataset,palette="husl",hue="sex")
plt.grid()
plt.show()
```

```python
sns.countplot(x="island",data=Dataset,palette="husl",hue="sex")
sns.countplot(x="species",data=Dataset,palette="husl",hue="sex")
plt.grid()
plt.show()
```
