```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
Dataset = sns.load_dataset("flights")
Dataset
```

```python
sns.barplot(x="year",y="passengers",data=Dataset,errorbar="sd")
plt.show()
```

```python
sns.barplot(x="year",y="passengers",data=Dataset,hue="month",errorbar="sd")
plt.show()
```
