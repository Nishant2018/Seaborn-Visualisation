```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
Dataset = sns.load_dataset("tips")
Dataset
```

```python
sns.violinplot(x="day",y="total_bill",data=Dataset)
plt.grid()
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=Dataset,hue="sex",split=False)
plt.grid()
plt.show()
```

```python
Dataset2 = sns.load_dataset("tips").head(30)
Dataset2
```

```python
sns.violinplot(x="day",y="total_bill",data= Dataset2,hue="sex",scale="count")
plt.grid()
plt.show()
```
