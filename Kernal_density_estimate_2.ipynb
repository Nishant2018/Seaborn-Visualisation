```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset("tips")
dataset
```

```python
sns.kdeplot(data= dataset)
plt.grid()
plt.show()
```

```python
sns.kdeplot(x="total_bill",data=dataset,linestyle="--")
sns.kdeplot(x="tip",data=dataset)
plt.grid()
plt.show()
```
